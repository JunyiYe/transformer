## Plot positional encoding

```python
import matplotlib.pyplot as plt
from models.embeddings.positional_encoding import PositionalEncoding 
```

```python
d_model = 512
max_len = 100
pe_layer = PositionalEncoding(d_model=d_model, max_len=max_len)

# Original pe buffer shape is (max_len, 1, d_model)
pe_tensor = pe_layer.pe.squeeze(1) # Shape: (max_len, d_model)

print(f"Tensor shape for plotting: {pe_tensor.shape}") # Should be (512, 100)

# --- Convert to NumPy array ---
pe_np = pe_tensor.numpy()

# --- Plotting ---
plt.figure(figsize=(10, 2))
im = plt.imshow(pe_np, aspect='auto', cmap='viridis')

plt.xlabel("Embedding Dimension Index")
plt.ylabel("Position in Sequence")
plt.title(f"Positional Encoding Matrix ({max_len} x {d_model})")
plt.colorbar(im, label="Encoding Value") # Add a color bar to show value scale

plt.show()
```
